# Number Guessing using while loop (infinte)
### Conditions
1. 7 trials
2. the number is betweem 1 to 100
3. if user entered invalid number then program should says "number is invalid"
4. if user number > actual number then, program should says " Your number is MAXIMUM "
5. if user number < actual number then,porgram should says "Your number is MINIMUM "

```python
secret_number=45
print("Hey,Welcome,")
print("Guess my number the range between 1 to 100")
for i in range(0,7):
    print("Trial: ",i+1)
    user_input=int(input("Enter the Number: "))
    if user_input < 0 or user_input > 100:
        print("Invalid number,Please enter the Number between 1 to 100")
    elif user_input>secret_number:
        print("Your number is Higher")
    elif user_input<secret_number:
        print("Your number is Low")
    else:
        print("Congratualtions!!🎊")
        break
```

# Same Concept for Login
### using for...else ,

Many beginners don't know this feature.
The else block runs only if the loop finishes without a break.

```python
password="Nandhitha"
login=False
for i in range(5):
    print("Trial",i+1)
    user_input=input("Enter the Password:")
    if password==user_input:
        print("Successfully! Login...")
        login=True
        break
    else:
        print("Wrong, Password, Try Again ,You loss Trial ",i+1,"Remaining Trials",5-(i+1))
if login!=True:
    print("Ooops......., Account is Blocked")
    print("Check, Your E-mail")
```
